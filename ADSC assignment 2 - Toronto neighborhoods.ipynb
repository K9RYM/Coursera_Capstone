Define the function to communicate with the API with 'get' requests

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    x=1
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        x+=1
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

How many unique categories can be curated from all the returned venues


```python
print('There are {} uniques categories.'.format(len(toronto_venues['Venue Category'].unique())))
toronto_dummies = pd.get_dummies(toronto_venues[['Venue Category']], prefix="", prefix_sep="")
toronto_dummies.head()
```

Adding Neighborhood column to toronto dummy values 

```python
toronto_dummies['Neighborhood'] = toronto_venues['Neighborhood']
fixed_columns = [toronto_dummies.columns[-1]] + list(toronto_dummies.columns[:-1])
fixed_columns.remove('Neighborhood')
a = ['Neighborhood']
new_list = a + fixed_columns
toronto_dummies = toronto_dummies[new_list]
```

```python
toronto_dummies.head()
```

Calculate the mean of the frequency of occurrence of each category

```python
toronto_dummies = toronto_dummies.groupby('Neighborhood').mean().reset_index()
toronto_dummies.head()
```

Define function to determine the top 10 venues for each neighborhood

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = toronto_dummies['Neighborhood']

for ind in np.arange(toronto_dummies.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_dummies.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

Now using K-means library to cluster each neighborhood of Toronto based on their venues

```python
# set number of clusters
kclusters = 5

toronto_clustering = toronto_dummies.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_
```

Add cluster labels to the top venues dataframe

```python
# add clustering labels
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

toronto_merged = toronto_nh

#add latitude/longitude for each neighborhood
toronto_merged = toronto_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighbourhood')

toronto_merged.head() # check the last columns!
```

Since there are some neighborhoods that Foursquare didnt return any venues for, they are not clustered by the algorithm, so we use .dropna function to drop the rows containing NaN values for cluster label

```python
toronto_merged = toronto_merged.dropna()
```

Now create a map of colored markers to plot the clustering of each neighborhood

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['Neighbourhood'], toronto_merged['Cluster Labels']):
    cluster = int(cluster)
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
